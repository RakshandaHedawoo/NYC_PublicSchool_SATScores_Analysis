Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# importing pandas for data manipulation
import pandas as pd

# Reading in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
#identifying the best schools based on their average math score being at least 80% of maximum score
score_criteria = 0.8*800
best_math_schools = schools[schools["average_math"] >= score_criteria][["school_name","average_math"]].sort_values("average_math",ascending=False)

#preview the result
best_math_schools.head()
```

```python
#Creating a new column to get the average total score
schools["total_SAT"]=schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools.head()

#subsetting data to just two columns in the resulting dataframe
best_performing_schools = schools[["school_name","total_SAT"]]
```

```python
#Arrange the schools in descending order to get high performing schools
top_10_schools = best_performing_schools.sort_values("total_SAT",ascending=False).head(10)
top_10_schools
```

```python
#importing numpy for aggregate functions
import numpy as np

#grouping data by borough
boroughwise_schools = schools.groupby("borough")["total_SAT"].agg([np.mean,np.std]).round(2)

#creating new column to find number of schools in each borough
boroughwise_schools["num_schools"] = schools.groupby("borough").size()

#renaming columns for more clarity in data
boroughwise_schools = boroughwise_schools.rename(columns={"mean":"average_SAT","std":"std_SAT"})

#locating the NYC borough with largest standard deviation
largest_std_dev = boroughwise_schools[boroughwise_schools["std_SAT"]==boroughwise_schools["std_SAT"].max()]

largest_std_dev
```
